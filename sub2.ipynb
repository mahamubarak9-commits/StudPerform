```python
# Simple Linear Regression
# How does writing score affect reading score?

# Step 1: Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step 2: Importing the dataset
dataset = pd.read_csv('StudentsPerformance.csv')
# Step 3: Independent (X) and Dependent (y) variables
# X = writing score, y = reading score
X = dataset.iloc[:, [7]].values      # writing score (column index 7)
y = dataset.iloc[:, [6]].values      # reading score (column index 6)

# Step 4: Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)

# Step 5: Training the Simple Linear Regression model on the Training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Step 6: Predicting the Test set results
y_pred = regressor.predict(X_test)

# Step 7: Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Reading Score vs Writing Score (Training set)')
plt.xlabel('Writing Score')
plt.ylabel('Reading Score')
plt.show()

# Step 8: Visualising the Test set results
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Reading Score vs Writing Score (Test set)')
plt.xlabel('Writing Score')
plt.ylabel('Reading Score')
plt.show()

# Step 9: Getting the final regression equation
print('Intercept:', regressor.intercept_)
print('Coefficient:', regressor.coef_)
```
